```python
# # 10
# empty = 4
# new_glass , remainder = 1, 1
# ans = 3 + 1 + (1) = 5


def get_num(num_empty):
  ans = 0
  while num_empty > 1:
    new_glass, remainder = divmod(num_empty, 3)
    ans += new_glass
    num_empty = remainder + new_glass
    if num_empty == 2:
      ans += 1
      break
  
  return ans

print(get_num(10))
```
