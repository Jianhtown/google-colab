### Problem 1 (5 points): 

Update your shopping list program from last week's homework to use a `while` statement, enabling users to add/remove as many items as they wish to their shopping list.

> 3 points: The user must be able to add items to and remove items from the list. The program must display the updated list after each add/remove.

> 1 point: the user must be able to tell the program to end

> 1 point: the program must prevent program errors from occuring due to bad user input

```python
# your code here
shopping_list = []
command = ""

while command.lower() != "end":
  print("please input your command : Add or Remove or End")
  command = input()
  try:
    if command.lower() == "add":
      print("\nFeel free to add any number items to your cart, but please separate each item in your list with a comma (',').  ")
      items = input().lower().split(",")
      for item in items:
        if item not in shopping_list:
          shopping_list.append(item)
      print("now your list is: ", shopping_list)
    elif command.lower() == "remove":
      print("\nFeel free to remove any number items to your cart, but please separate each item in your list with a comma (',').  ")
      items = input().lower().split(",")
      for item in items:
        if item in shopping_list:
          shopping_list.remove(item)
      print("now your list is: ", shopping_list)
    elif command.lower() == "end":
      print("now your list is: ", shopping_list,"bye!")
      break
    else:
      print("wrong input, type again")
  except:
    print("WRONG input, type again")
  
```
