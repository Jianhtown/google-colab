### Problem One (4 points): 
Write a program that converts a Celsius temperature to Farhenheit and vice-versa. 

> 3 points: Your program must ask a user to input a temperature and a measurement system, and then tell the user the converted temperature. Your program must also handle situations when users enter incorrect information.

> 1 point: Explain, step-by-step, what your solution for Problem \#1 is doing by adding comments in your code above.

```python
# your code here
#expecting an input with correct format described below, upper and lower cases are all accpetable.
print("Enter the temprature with unit C or F following the number:")
temp = input()
#slicing the input string into the unit and number, and convert the string into integer to perform further conversions.
temp_unit = temp[-1]
temp_number = int(temp[:-1])
#the control flow testing if the input unit capitalized equals "C" or "F", and applying related conversion functions to the number.
if temp_unit.upper() == "C":
  result = round((9 * temp_number) / 5 + 32) #rounded up, converting from Celsius to Farhenheit
  print(temp, "equal", result, "F.")
elif temp_unit.upper() == "F":
  result = round((temp_number - 32) * 5 / 9) #rounded up, converting from Fahrenheit to Celcius
  print(temp, "equal", result, "C.")
#catching errors
else:
  print("wrong input.")
```
